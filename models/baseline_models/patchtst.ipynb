```python
import sys
import torch
from torch.utils.data import Dataset
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
from nbeats_pytorch.model import NBeatsNet
from sklearn.preprocessing import MinMaxScaler
import torch.optim as optim

base_path = '/home/noam.koren/multiTS/NFT/'
sys.path.append(base_path)

from dicts import data_to_num_vars_dict
from models.training_functions import read_all_data_and_print_stats, train_model, evaluate_model, save_model
from models.baseline_models.base_models import FC, LSTM, CNN, TCN, TimeSeriesTransformer, PatchTST

data = 'ett'

n_vars = data_to_num_vars_dict[data]
horizon = 1
lookback = 5
batch_size = 32
plot_epoch = 5
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    
class dataset(Dataset):
    def __init__(self, data, target):
        self.input = data
        self.target = target

    def __len__(self):
        return len(self.input)

    def __getitem__(self, idx): 
        return self.input[idx], self.target[idx]


if data == 'ecg':
    data_path = f"{base_path}data/{data}/{data}_{lookback}l_{horizon}h_{n_people}people_{n_rows}rows/"
else:
    data_path = f"{base_path}data/{data}/{data}_{lookback}l_{horizon}h/"

train_X, train_y, val_X, val_y, test_X, test_y = read_all_data_and_print_stats(data_path=data_path, print_stats=False)
```

```python
model = PatchTST(n_vars, lookback, horizon).to(device)

train_model(
    dataset=dataset, 
    epochs=10, 
    model=model, 
    train_X=train_X.to(device), train_y=train_y.to(device),
    val_X=val_X.to(device), val_y=val_y.to(device), 
    device=device, 
    lookback=lookback,
    horizon=horizon, 
    n_vars=n_vars, 
    batch_size=batch_size, 
    print_epoch=plot_epoch, 
    path_to_save_prediction_plots=None,
    )

evaluate_model(model, train_X.to(device), train_y.to(device), val_X.to(device), val_y.to(device), test_X.to(device), test_y.to(device))
```
